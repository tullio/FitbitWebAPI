```scala
# Compute x^2 + y^2 across a 2D grid
x, y = np.meshgrid(range(-5, 5), range(-5, 5))
z = x ** 2 + y ** 2

# Convert this grid to columnar data expected by Altair
source = pd.DataFrame({'x': x.ravel(),
                     'y': y.ravel(),
                     'z': z.ravel()})

alt.Chart(source).mark_rect().encode(
    x='x:O',
    y='y:O',
    color='z:Q'
)
```

```scala


import me.shadaj.scalapy.interpreter.CPythonInterpreter
val sys = py.module("sys")
CPythonInterpreter.execManyLines(
  """class OutputRedirector:
    |  def __init__(self, write_fn, flush_fn):
    |    self.write_fn = write_fn
    |    self.flush_fn = flush_fn
    |  def write(self, text):
    |    self.write_fn(text)
    |  def flush(self):
    |    self.flush_fn()""".stripMargin
)

val stdoutRedirector = py.Dynamic.global.OutputRedirector(
  py.Any.from((text: String) => {
    System.out.print(text)
    py.None
  }),
  py.Any.from(() => {
    System.out.flush()
    py.None
  })
)

sys.stdout = stdoutRedirector

val stderrRedirector = py.Dynamic.global.OutputRedirector(
  py.Any.from((text: String) => {
    System.err.print(text)
    py.None
  }),
  py.Any.from(() => {
    System.err.flush()
    py.None
  })
)

sys.stderr = stderrRedirector

```
